```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest')
```

**INSIGHTS FOR EDA**

The strongest positive correlation occurs among "Bank Deposits" with "Checking Accounts", "Saving Accounts", and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.

**DEMOGRAPHIC INSIGHTS**  

**Age**: Clients range from 17 to 85 years, with an average age of 51.

**Nationality**:

- European: 1,309 clients (largest group)

- Asian: 754

**RISK WEIGHTING**

Most clients are low-to-medium risk (1,222 clients have rating 2; only 160 have rating 5).

**CORRELATION PATTERNS**

**ESTIMATED INCOME ↔ RISK WEIGHTING (0.66)**

Meaning: Higher-income clients tend to be assigned higher risk scores.

Business Impact: These clients may be taking on more complex financial products, larger loans, or riskier investments.

While they generate more revenue potential (through interest, fees, and cross-sells), they also require stricter risk management to avoid defaults.

Banks might need premium risk monitoring services or custom credit terms for this segment.

**ESTIMATED INCOME ↔ SUPERANNUATION SAVINGS (0.37)**

Meaning: Higher earners save more in retirement accounts.

Business Impact: This segment is ideal for wealth management services — pension planning, investment portfolios, insurance packages.

Offering exclusive retirement products to high-income customers could boost long-term customer retention and fee-based revenue.

**ESTIMATED INCOME ↔ BANK LOANS (0.33) & BUSINESS LENDING**

Meaning: Higher-income clients are also borrowing more.

Business Impact: Suggests a profile of financially active clients who leverage credit to expand businesses or invest. These clients represent high-value lending opportunities but also require robust credit scoring models to balance profitability with risk.

Potential to cross-sell insurance products, loan protection plans, and business advisory services.

**LOW CORRELATION WITH DEMOGRAPHICS (AGE, GENDER, LOCATION ID)** 

Meaning: Income and financial activity aren’t strongly tied to age, gender, or location in this dataset.

Business Impact: Targeting should be behavior-based (financial habits, product usage) rather than purely demographic.

Marketing can be personalized by financial patterns — e.g., clients with high borrowing and high deposits — instead of broad age or gender segments.
